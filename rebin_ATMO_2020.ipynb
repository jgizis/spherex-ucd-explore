```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import glob
import spectres
```

```python
from astropy.io import ascii
from astropy import units as u
from astropy.visualization import quantity_support
from astropy.table import Table
#dont actually use units
```

```python
x_values=np.load('spiff_xbins.npy')
#note that both SPIFF/SphereX and these models are in microns 
```

```python
#test on a sample spectrum
f='/Users/gizis/model_atmospheres/ATMO_2020/atmosphere_models/NEQ_weak_spectra/spec_T800_lg4.0_NEQ_weak.txt'
t = Table.read(f,format='ascii',data_start=3)
yy=spectres.spectres(x_values,t['col1'],t['col2'])
new_name = f.replace('/spec_','spherex_')
outname= new_name.replace('.txt','.csv')
print(outname)
newtable=Table([x_values,yy],names=('wavelength_um','flux_flambda'))
newtable
```

```python
#plot the test
plt.plot(t['col1'],t['col2'])
plt.plot(newtable['wavelength_um'],newtable['flux_flambda'])
plt.xlim(0.6,6.0)
```

```python
file_paths = glob.glob('/Users/gizis/model_atmospheres/ATMO_2020/atmosphere_models/NEQ_weak_spectra/spec*.txt')
#
for f in file_paths:
    t = Table.read(f,format='ascii',data_start=3)
    yy=spectres.spectres(x_values,t['col1'],t['col2'])
    new_name = f.replace('spectra/spec_','spectra/spherex_')
    outname= new_name.replace('.txt','.csv')
    print(outname)
    newtable=Table([x_values,yy],names=('wavelength_um','flux_flambda'))
    newtable.write(outname,overwrite=True)
    del newtable
```
